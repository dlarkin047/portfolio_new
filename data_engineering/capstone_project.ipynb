## Class and functions

```python
class RetailSalesAnalyser:
    def __init__(self):
        self.data = pd.read_csv('dataset/retail_sales.csv')
        self.data['Date'] = pd.to_datetime(self.data['Date'])

    def clean_date(self):
        self.data.dropna(inplace=True)

    def total_sales_per_product(self):
        return self.data.groupby(['Product'], as_index = False)['Sales'].sum()
    
    def best_seller(self):
        return self.data.groupby('Product', as_index = False)['Quantity'].sum()['Product'].max()
    
    def daily_average(self):
        return self.data.groupby(['Date'], as_index = False)['Sales'].mean()

    def plot_sale_trends(self):
        self.data.groupby('Date')['Sales'].sum().plot(kind='line')
        plt.title('Sales by Date')
        plt.xlabel('Date')
        plt.ylabel('Sales')
        plt.show()

    def plot_sales_per_product(self):
        self.total_sales_per_product().plot(x='Product', y='Sales', kind='bar', title='Sales per Product')
        plt.show()
```

```python
analyser = RetailSalesAnalyser()
print('Total sales per Product:\n',analyser.total_sales_per_product())
print('\nBest Seller:\n',analyser.best_seller())
print('\nDaily Average:\n', analyser.daily_average())

analyser.plot_sale_trends()
analyser.plot_sales_per_product()
```
